```
# 需要修改txt文件
with open('categories1.txt', 'r') as f:
  categories = f.readlines()
```

```
categories[:5]
```

```
categories = [c.replace('\n', '') for c in categories]
```

```
categories[:5]
```

```
import tensorflow as tf
```

```
def download(categories):
  count = 0
  base = 'https://storage.googleapis.com/quickdraw_dataset/full/numpy_bitmap/'
  for c in categories:
    path = '/content/dataset/' + c + '.npy'
    category = c.replace(' ', '%20')
    url = base + category + '.npy'
    tf.keras.utils.get_file(fname=path, origin=url)
    count += 1
  print('{} categories have been downloaded successfully.'.format(count))
```

```
download(categories)
```
